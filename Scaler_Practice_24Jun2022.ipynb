```python
def sigmoid(x):
  return 1/(1+exp(-x))
```

```python
vectorized_sigmoid=np.vectorize(sigmoid)
```

```python
vectorized_sigmoid([-2,-1,0,1,1])
```

```python
x=np.arange(-10,10,0.1)
y=vectorized_sigmoid(x)
y
```
