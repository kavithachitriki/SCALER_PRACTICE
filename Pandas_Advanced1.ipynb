```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("movies.csv",index_col=0)
movies.head()
```

```python
directors=pd.read_csv("directors.csv",index_col=0)
directors.head()
```

```python
movies.shape
```

```python
directors.shape
```

```python
movies["director_id"].nunique()
```

```python
directors["id"].nunique()
```

```python
np.all(movies["director_id"].isin(directors["id"]))
```

```python
unique_movies=movies["director_id"].unique()
unique_directors=directors["id"].unique()
np.all(unique_movies.isin(unique_directors))
```

```python
data=movies.merge(directors,how="left",left_on="director_id",right_on="id")
data.head()
```

```python
data["title"].nunique()==data.shape[0]
```

```python
data.drop(["director_id","id_x","id_y"],axis=1,inplace=True)
```

```python
data["revenue"]=(data["revenue"]/1000000).round(2)
```

```python
data["budget"]=(data["budget"]/1000000).round(2)
```

```python
df_agg=data.groupby("director_name")[["title","year"]].aggregate({"year":["min","max"],"title":"count"})
df_agg
```

```python
df_agg.columns
```

```python
df_agg["title"]
```

```python
df_agg.columns=["_".join(col) for col in df_agg.columns]
```

```python
df_agg
```

```python
df_agg.reset_index(inplace=True)
df_agg
```

```python
df_agg["yrs_active"]=df_agg["year_max"]-df_agg["year_min"]
```

```python
df_agg["movies_per_year"]=df_agg["title_count"]/df_agg["yrs_active"]
```

```python
df_agg
```

```python
df_agg.sort_values("movies_per_year",ascending=False)
```

```python
data["is_risky"]=data.groupby("director_name").apply(lambda x:x["budget"]-x["revenue"].mean()).values>0
```

```python
def func(x):
  x["risky"]=(x["budget"]-x["revenue"].mean())>0
  return x
data=data.groupby("director_name").apply(func)
```
