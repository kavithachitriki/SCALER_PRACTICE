```python
x=np.arange(-3,3.1,0.1)
y=x**2
plt.plot(x,y,color="k")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Squares")
plt.xlim(-15,15)
plt.ylim(0,11)
plt.show()
```

```python
x=np.arange(-3,3.1,0.1)
y=x**2
y1=x**3
plt.plot(x,y,color="r",label="Squares")
plt.plot(x,y1,color="b",label="Cubes")
plt.legend(loc="lower right",ncol=2,frameon=False)
plt.text(0,0,"(0,0)")
plt.text(1,1,"(1,1)")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Squares")
plt.grid()
plt.show()
```

```python
x=np.arange(-3,3.1,0.1)
y=x**2
y1=x**3
plt.plot(x,y,color="r",label="Squares")
plt.plot(x,y1,color="b",label="Cubes")
plt.legend(loc="lower right",ncol=2,frameon=False,bbox_to_anchor=(1.5,0.5))
plt.text(0,0,"(0,0)")
plt.text(1,1,"(1,1)")

plt.title("Test Graph")
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
y1=np.cos(x)

plt.plot(x,y)
plt.show()
plt.plot(x,y1)
plt.show()
```

```python
plt.figure(figsize=(1,1))
fig,ax=plt.subplots(3,1,figsize=(16,4))
y2=np.tan(x)
ax[0].plot(x,y)

ax[1].plot(x,y1)
ax[2].plot(x,y2)
plt.show()
```
